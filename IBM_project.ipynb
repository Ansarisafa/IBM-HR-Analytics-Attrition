# **Import the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Exploratory Data Analysis(EDA)

📊 Monthly Income & Age Distribution

```python
sns.boxplot(x='Attrition', y='MonthlyIncome',color='c', data=df)
plt.title('Income vs Attrition')
plt.show()

sns.histplot(data=df, x='Age', hue='Attrition',color='c', kde=True, bins=30)
plt.title('Age Distribution by Attrition')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
sns.histplot(data=df, x='Age', kde=True,color='c', ax=axes[0])
axes[0].set_title('Distribution Employee by Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')

sns.countplot(data=df, x='Gender',color='c', ax=axes[1])
axes[1].set_title('Distribution Employee by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Count')

sns.countplot(data=df, x='Department',color='c', ax=axes[2])
axes[2].set_title('Distribution Employee by Department')
axes[2].set_xlabel('Department')
axes[2].set_ylabel('Count')
plt.tight_layout()
plt.show()
```

1. Age: Most of the company's employees are in the 30-35 age group. This
indicates that the company has many employees who are at a productive and
experienced age.
2. Gender: The majority of employees at this company are male. There are
significantly more male employees than female employees.
3. Department: Most of the company's employees are concentrated in the
research and development department. This indicates that the company is
heavily focused on product or service research and development activities.

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Department', hue='Gender')
plt.title('Distribution of Employees by Department and Gender')
plt.xticks(rotation=45)
plt.show()
```

# **Correlation Analysis**

Correlation analysis is used to measure the strength and direction of the relationship between two or more numeric variables. In this project, it helps us identify which factors (like age, income, or working hours) are most related to employee attrition.

```python
plt.figure(figsize=(15,10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
